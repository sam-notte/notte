```python
import asyncio
from notte.env import NotteEnv
from examples.agent import SimpleNotteAgent
import json
from notte.password.vault import Vault

vault = Vault()
vault.add_vault(json.load(open("vault_path")))

# No need to explicitly get credentials anymore
agent = SimpleNotteAgent(
    model="openai/gpt-4o",
    max_steps=10,
    headless=False
)

# Run the agent directly with await
task = "Post 'Hello World' on Twitter"
result = await agent.run(
    task=task,
    url="https://x.com"  # Credentials will be loaded automatically
)
```
